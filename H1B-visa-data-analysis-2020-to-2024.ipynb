This is a notebook of Exploratory data analysis of H1B visa from 2020 to the year 2024. 

About the Dataset
This dataset provides a comprehensive record of Labor Condition Application (LCA) disclosures for H1B visa petitions filed with the U.S. Department of Labor (DOL) from 2020 to 2024. The H1B visa is a non-immigrant visa that allows U.S. companies to employ foreign workers in specialty occupations requiring theoretical or technical expertise. These roles typically include fields such as IT, engineering, finance, healthcare, and more. The H1B program is critical for addressing skill gaps in the U.S. workforce and supporting economic growth.

Key features
Years Covered: 2020 to 2024.
Number of Records: Includes all LCAs filed during this period.
Attributes: Comprehensive fields such as case number, job title, employer name, wage data, work location, and visa classification.
Source: Data is based on public disclosures from the U.S. Department of Labor.

Source of Dataset 
https://www.kaggle.com/datasets/zongaobian/h1b-lca-disclosure-data-2020-2024?select=Combined_LCA_Disclosure_Data_FY2020_to_FY2024.csv

```python
# Handling missing values

h1b_visa_data_2020_to_2024.fillna(h1b_visa_data_2020_to_2024.median(numeric_only=True),inplace=True)
```

```python
column_list = h1b_visa_data_2020_to_2024.columns.tolist()

for column_list in column_list:
    print(column_list,h1b_visa_data_2020_to_2024[column_list].isnull().sum())
   
```

```python
# Dropping null rows

h1b_visa_data_2020_to_2024 = h1b_visa_data_2020_to_2024.dropna(subset=["JOB_TITLE", "EMPLOYER_NAME", "EMPLOYER_STATE", "WORKSITE_STATE"])
```

From 2020 to 2024, over 2.76 million H1B petitions were certified which means 92.75%, vastly outnumbering withdrawn (53,741) or denied (17,585) cases, highlighting the program’s high approval rate. The dominance of "Certified" statuses suggests strong employer compliance with visa requirements, while low denial rates reflect rigorous pre-filing vetting to meet USCIS standards.

```python
# Distribution of Prevailing wage with Histogram plot

plt.figure(figsize=(10,6))
sns.histplot(h1b_visa_data_2020_to_2024_certified["PREVAILING_WAGE"],bins=30,kde=True)
plt.title("Distribution of wage rate among the certified candidate")
plt.show()

```

From 2020 to 2024, the H1B prevailing wage distribution shows most certified candidates earned mid-range salaries (e.g., ~$100,000), with fewer individuals in extremely high or low wage brackets. The pattern aligns with typical H1B trends, where tech and engineering roles dominate, and employers adhere to prevailing wage requirements to ensure fair compensation.

```python
# Distribution of worksite states

top_worksite_state = h1b_visa_data_2020_to_2024_certified["WORKSITE_STATE"].value_counts().head(10)
```

```python
# plot bar graph for worksite states

top_worksite_state.plot(kind="bar",title="Top 10 Worksite States",figsize=(10,6),colormap='coolwarm')
plt.show()
```

```python
# Average wage by states

average_wage_by_state = h1b_visa_data_2020_to_2024_certified.groupby("WORKSITE_STATE")["WAGE_RATE_OF_PAY_FROM"].mean().sort_values(ascending=False)
print(average_wage_by_state)
```

```python
# plot bar graph of top 10 average wage by state 
average_wage_by_state.head(10).plot(kind="bar",title="Top 10 States by average wages",figsize=(10,6),colormap='rainbow')
plt.show()
```

The bar chart presents the top 10 states with the highest average wages for H1B visa applications from 2020 to 2024. West Virginia leads with the highest average wage, followed by Washington and California. The remaining states, including Maine, New York, Massachusetts, Mississippi, Delaware, District of Columbia, and New Jersey show a gradual decrease in average wages.

```python
# plot bar graph of low 10 average wage by state 
average_wage_by_state.tail(10).plot(kind="bar",title="Low 10 States by average wages",figsize=(10,6),colormap='rainbow')
plt.show()
```

The bar chart presents the bottom 10 states with the lowest average wages for H1B visa applications from 2020 to 2024. Michigan has the lowest average wage, followed by Vermont and Wyoming. The remaining states, including North Dakota, Alaska, Puerto Rico, Palau, Guam, Virgin Islands, and Northern Mariana Islands, show a gradual increase in average wages.

```python
# Remote work trends

h1b_visa_data_2020_to_2024_certified["RECEIVED_DATE"] = pd.to_datetime(h1b_visa_data_2020_to_2024_certified["RECEIVED_DATE"], errors='coerce')
h1b_visa_data_2020_to_2024_certified['YEAR'] = h1b_visa_data_2020_to_2024_certified["RECEIVED_DATE"].dt.year

remote_trends = h1b_visa_data_2020_to_2024_certified[h1b_visa_data_2020_to_2024_certified["WORKSITE_STATE"]!=h1b_visa_data_2020_to_2024_certified["EMPLOYER_STATE"]]
remote_trends_by_year = remote_trends.groupby('YEAR')['CASE_NUMBER'].count()

print(remote_trends_by_year)
```

The H1B remote work trend surged sharply from 61,459 cases in 2019 to 311,262 in 2020 (a 406% increase), likely driven by pandemic-era remote work adoption. While numbers declined gradually post-2020 (down to 201,043 in 2024), they remained 3-5x higher than pre-pandemic levels, reflecting a lasting shift toward hybrid/remote work in H1B-dependent sectors like tech and consulting.

```python
# Plotting the remote trends

plt.figure(figsize=(10,6))
sns.lineplot(data=remote_trends_by_year,marker='o',color='green')
plt.title("Remote trends by year")
plt.show()

```
