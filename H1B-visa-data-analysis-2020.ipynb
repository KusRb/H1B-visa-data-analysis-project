This is a notebook of Exploratory data analysis of H1B visa for the year 2020. 

About the Dataset
This dataset provides a comprehensive record of Labor Condition Application (LCA) disclosures for H1B visa petitions filed with the U.S. Department of Labor (DOL) 2020. The H1B visa is a non-immigrant visa that allows U.S. companies to employ foreign workers in specialty occupations requiring theoretical or technical expertise. These roles typically include fields such as IT, engineering, finance, healthcare, and more. The H1B program is critical for addressing skill gaps in the U.S. workforce and supporting economic growth.

Key features
Years Covered: 2020
Number of Records: Includes all LCAs filed during this period.
Attributes: Comprehensive fields such as case number, job title, employer name, wage data, work location, and visa classification.
Source: Data is based on public disclosures from the U.S. Department of Labor.

Source of Dataset 
https://www.kaggle.com/datasets/zongaobian/h1b-lca-disclosure-data-2020-2024/data?select=Combined_LCA_Disclosure_Data_FY2020.csv

```python
# Handling missing values

h1b_visa_data_2020.fillna(h1b_visa_data_2020.median(numeric_only=True),inplace=True)
```

```python
column_list = h1b_visa_data_2020.columns.tolist()

for column_list in column_list:
    print(column_list,h1b_visa_data_2020[column_list].isnull().sum())
   
```

```python
# Dropping null rows

h1b_visa_data_2020 = h1b_visa_data_2020.dropna(subset=["JOB_TITLE", "EMPLOYER_NAME"])
```

```python
# Checking for duplicates 

h1b_visa_data_2020.duplicated().sum()
```

```python
# Counting the values of each case status

h1b_visa_data_2020['CASE_STATUS'].value_counts()
```

In 2020, there were 545616 certified applicant which means 94.50% of the total applicant were approved. Similarly, 16738 were approved initially, but the employer later decided to withdraw it. 10991 withdrawn their application whereas only 3983(0.69%) were denied. 

```python
# Filtered dataset for certified case only 

h1b_visa_data_2020_certified = h1b_visa_data_2020[h1b_visa_data_2020['CASE_STATUS']=='Certified']
```

```python
# Distribution plot of top 10 Job Status
top_jobs = h1b_visa_data_2020_certified["JOB_TITLE"].value_counts().head(10)
```

From the above data, mostly the IT professionals are highly certified such as Software Engineer, Software Developer, and Senior System Analyst.


```python
# Distribution of Prevailing wage with Histogram plot

plt.figure(figsize=(10,6))
sns.histplot(h1b_visa_data_2020_certified["PREVAILING_WAGE"],bins=30,kde=True)
plt.title("Distribution of wage rate among the certified candidate")
plt.show()

```

The histogram shows that the majority of certified H-1B visa candidates in 2020 were employed at lower wages, with a smaller portion receiving much higher wages. This indicates a wide range of salaries across different occupations and skill levels within the H-1B program. The most common wage rate appears to be around $100,000. This is where the histogram has its tallest bar.

```python
# Distribution of worksite states

top_worksite_state = h1b_visa_data_2020_certified["WORKSITE_STATE"].value_counts().head(10)
```

```python
# plot bar graph for worksite states

top_worksite_state.plot(kind="bar",title="Top 10 Worksite States",figsize=(10,6),colormap='coolwarm')
plt.show()
```

The bar plot shows the top 10 states with the highest number of certified H-1B visa applications in 2020. California has the highest number of certified H-1B applications, followed by Texas and New York. The remaining states in the top 10 are Washington, New Jersey, Illinois, Massachusetts, North Carolina, Pennsylvania, and Georgia.

```python
# Average wage by states

average_wage_by_state = h1b_visa_data_2020_certified.groupby("WORKSITE_STATE")["WAGE_RATE_OF_PAY_FROM"].mean().sort_values(ascending=False)
print(average_wage_by_state)
```

```python
# plot bar graph of top 10 average wage by state 
average_wage_by_state.head(10).plot(kind="bar",title="Top 10 States by average wages",figsize=(10,6),colormap='rainbow')
plt.show()
```

The bar plot illustrates the top 10 states with the highest average wages for certified H-1B visa candidates in 2020. West Virginia ranks first with the highest average wage, followed by Washington and California. The remaining states in the top 10 are Texas, Mississippi, Maine, New York, Montana, North Dakota, and South Dakota.

```python
# plot bar graph of low 10 average wage by state 
average_wage_by_state.tail(10).plot(kind="bar",title="Low 10 States by average wages",figsize=(10,6),colormap='rainbow')
plt.show()
```

The bar plot illustrates the bottom 10 states with the lowest average wages for certified H-1B visa candidates in 2020. Tennessee has the lowest average wage, followed by Kentucky and Indiana. The remaining states in the bottom 10 are Alaska, Michigan, Puerto Rico, Palau, Guam, the Virgin Islands, and the Northern Mariana Islands.

```python
# Remote work trends

h1b_visa_data_2020_certified["RECEIVED_DATE"] = pd.to_datetime(h1b_visa_data_2020_certified["RECEIVED_DATE"], errors='coerce')
h1b_visa_data_2020_certified['YEAR'] = h1b_visa_data_2020_certified["RECEIVED_DATE"].dt.year

remote_trends = h1b_visa_data_2020_certified[h1b_visa_data_2020_certified["WORKSITE_STATE"]!=h1b_visa_data_2020_certified["EMPLOYER_STATE"]]
remote_trends_by_year = remote_trends.groupby('YEAR')['CASE_NUMBER'].count()

print(remote_trends_by_year)
```

The data shows a significant increase in remote work trends between 2019 and 2020, with the number of cases rising from 61,459 in 2019 to 217,708 in 2020. This sharp rise likely reflects the impact of the COVID-19 pandemic, which accelerated the adoption of remote work practices across various industries.
